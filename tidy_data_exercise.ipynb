```python
import pandas as pd
```

```python
def get_attendance_grade(letter):
    if letter == 'P':
        return 1
    elif letter == 'A':
        return 0
    elif letter == 'T':
        return 0.9
    elif letter == 'H':
        return 0.5

get_attendance_grade('H')
```

```python
grades = {
    'P': 1,
    'A': 0,
    'H': 0.5,
    'T': 0.9,
}

df = pd.read_csv('untidy-data/attendance.csv')
df = df.rename(columns={'Unnamed: 0': 'student'})
df = df.melt(id_vars='student', var_name='date')
df['grade'] = df.value.apply(get_attendance_grade)
df.groupby('student').grade.mean()
```

```python
df = pd.read_csv('untidy-data/coffee_levels.csv')
df = df.pivot_table(values='coffee_amount', index='hour', columns='coffee_carafe')
df.round(3)
```

```python
df = pd.read_csv('untidy-data/cake_recipes.csv')
recipes_and_positions = df['recipe:position'].str.split(':', expand=True)
recipes_and_positions.columns = ['recipe', 'position']
df = pd.concat([df, recipes_and_positions], axis=1).drop(columns='recipe:position')
df = df.melt(id_vars=['recipe', 'position'], var_name='temp', value_name='tastiness')
df
```

Which recipe, on average, is the best? recipe b

```python
df.groupby('recipe').tastiness.mean()
```

Which oven temperature, on average, produces the best results? 275

```python
df.groupby('temp').tastiness.mean()
```

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degree

```python
df.sort_values(by='tastiness').tail()
```
